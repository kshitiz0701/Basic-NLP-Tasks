## Background: What is Zipf's Law?

Zipfs Law states that in a large corpus of natural language, the frequency of any word is inversely proportional to its rank in the frequency table. This implies that frequency and rank are related by a power law.


Zipf's Law can be expressed as:
$$
f(r) = \frac{C}{r^k}
$$
Where:
- \( f(r) \) is the frequency of the word at rank \( r \).
- \( r \) is the rank of the word (1 for the most frequent word, 2 for the second most frequent, etc.).
- \( C \) is a constant that depends on the distribution of the frequencies.
- \( k \) is a parameter that decides the power law relationship.

For example, for some given constant 'k'

- The most frequent word will appear roughly twice as often as the second most frequent word.
- The second most frequent word will appear approximately three times as often as the third most frequent word, and so on.

## What are Regular Expressions?
Regular expressions (regex or regexp) (https://en.wikipedia.org/wiki/Regular_expression) are powerful tools for finding or matching patterns in strings. They allow you to define specific patterns and search for characters or words within text.


## Regular expressions in Python
Python has a built-in package called `re` (https://docs.python.org/3/library/re.html), which can be used to work with Regular Expressions. Using this module, you specify the rules for the set of possible strings that you want to match; this set might contain English sentences, or e-mail addresses, or TeX commands, or anything you like. You can then ask questions such as Does this string match the pattern?, or Is there a match for the pattern anywhere in this string?. You can also use REs to modify a string or to split it apart in various ways.

## Rules

Regular expressions can contain both special and ordinary characters. Most ordinary characters, like `'A'`, `'a'`, or `'0'`, are the simplest regular expressions; they simply match themselves. You can concatenate ordinary characters, so `last` matches the string `'last'`.

Some characters, like `'|'` or `'('`, are special. Special characters either stand for classes of ordinary characters, or affect how the regular expressions around them are interpreted.

```python
import pandas as pd
import re


#  regex pattern for finding the asked informations

words_regex = r'\b[a-zA-Z]+\b'
#  Here \b is indicating the start of a word
# [a-zA-Z] is indicating a string of words containing only alphabets in small and capital format
# +\b is indicating the end of the word
# It identifies the start and the end of the word by the help of white spaces between the words

emails_regex = r'[a-zA-Z0-9_.%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{,4}'
# Here [a-zA-Z0-9_.%+-]+ is indicating oer or more character continuous word which can contain alphabets both in small and capital form it can also contain _,.,%,+,-
# Then it is followed by a @ symbol
# and then a word with any alphanumerical character including . and -
#  then comes the dot and then a string of words that only contain alphabets and I am counting only 4 of it at max

phone_regex = r'\(?\d{3}\)?[- ]?\d{3}[- ]?\d{4}'
#   I have considered all the patterns whether it is (XXX)XXXXXXXXX or in xxx-xxx-xxxx format and if brackets are there or not

urls_regex = r'https?://[^\s]+'
# it will capture all the things which are starting from http or https


# Functions to apply regex and extract information
def to_find_words(doc):
    return re.findall(words_regex, doc)
def to_find_emails(doc):
    return re.findall(emails_regex, doc)
def to_find_phones(doc):
    return re.findall(phone_regex, doc)
def to_find_urls(doc):
    return re.findall(urls_regex, doc)

batch_size = 5000

def batch_processing(datafr, st, en):
    batch = datafr.iloc[st:en].copy()
    batch['words']= batch['message'].apply(to_find_words)
    batch['emails'] = batch['message'].apply(to_find_emails)
    batch['phones'] = batch['message'].apply(to_find_phones)
    batch['urls'] = batch['message'].apply(to_find_urls)
    return batch

# I am doing batch processing because it was taking very much time to find all the words and my system crashes because of it so I have done batch processing and have saved the
# results and then i am going to next batch. I have referred to pandas documentation for learning about batch processing

# Citations
# 1. https://www.w3schools.com/python/python_regex.asp
# 2. used pythex for checking the regex patterns
# 3. Pandas Documentation
```

```python

import gc
number_of_batch = 0

for i, st in enumerate(range(0, len(dataframe), batch_size)):
    en = min(st + batch_size, len(dataframe))
    processed_batch = batch_processing(dataframe, st, en)
  # Saving the processed batch into the memory so that my system does not crashes
    processed_batch.to_parquet(f'processed_batch_{i}.parquet')
    number_of_batch = number_of_batch+1  # I have to keep track of how many batches are saved in total as I have to read the batches further

    # Freeing the memory
    del processed_batch
    gc.collect()

#  reading all the saved batches one by one
dfs = [pd.read_parquet(f'processed_batch_{i}.parquet') for i in range(number_of_batch)]
# concatinating the results in the dataframe
final_df = pd.concat(dfs, ignore_index=True)

#  this cell takes 6 mins to run
#  Citations
#  1. https://docs.python.org/3/library/gc.html for gc
```

```python
from collections import Counter
import numpy as np
import pandas as pd

# Function to convert items to strings and flatten nested structures
def convert_to_strings(items):
    if isinstance(items, (list, tuple, np.ndarray)):
        # Flatten nested lists/tuples and convert all elements to strings
        return [str(item) for sublist in items for item in (sublist if isinstance(sublist, (list, tuple)) else [sublist])]
    return [str(items)]  # Returns single items as a list of strings

# Applied the conversion function across specified columns in the DataFrame
def apply_conversion(df, columns):
    for column in columns:
        df[column] = df[column].apply(convert_to_strings)
    return df

# Converted the necessary columns in the DataFrame
final_df = apply_conversion(final_df, ['phones', 'emails', 'urls', 'words'])

# Function to count occurrences of items in a specified column
def count_occurrences(df, column):
    return Counter([item for sublist in df[column] for item in sublist])

# Count the occurrences for each category
word_counts = count_occurrences(final_df, 'words')
email_counts = count_occurrences(final_df, 'emails')
phone_counts = count_occurrences(final_df, 'phones')
url_counts = count_occurrences(final_df, 'urls')

# I am using the most common function to get the top 200 most common entries for each category
top_200_words = word_counts.most_common(200)
top_200_emails = email_counts.most_common(200)
top_200_phones = phone_counts.most_common(200)
top_200_urls = url_counts.most_common(200)

# Citations
# 1. https://www.geeksforgeeks.org/python-counter-objects-elements/ for counter function
# 2. https://www.geeksforgeeks.org/python-most_common-function/
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to plot a histogram
def plot(data, title, xlabel, ylabel, top_n=200):

    # Separate the elements and their counts
    if len(data) > top_n:
        elements, counts = zip(*data[:top_n])
    else:
        elements, counts = zip(*data)

    plt.figure(figsize=(14, 7))  # Adjusted figure size for better visibility
    sns.barplot(x=list(elements), y=list(counts), palette="rainbow")

    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.tight_layout()  # Adjust layout to make room for x-axis labels
    plt.show()

# Plotting histograms for each category

# plotting top 200 Words by their frequency
plot(top_200_words, 'Top 200 Words by Frequency', 'Words', 'Frequency')

# plotting top 200 Emails by their frequency
plot(top_200_emails, 'Top 200 Emails by Frequency', 'Emails', 'Frequency')

# plotting top 200 phone numbers by their frequency
plot(top_200_phones, 'Top 200 Phone Numbers by Frequency', 'Phone Numbers', 'Frequency')

# plotting top 200 urls by their frequency
plot(top_200_urls, 'Top 200 URLs by Frequency', 'URLs', 'Frequency')

# Citations
# 1.https://matplotlib.org/ for plotting
```

# **Question 4: N-Gram models**

Language models are statistical models that give out the probability distribution of a sentence (sequence of tokens). They rely on the joint probability distribution of the tokens in the sentence.

For example, let us calculate the probability of the sentence "Monza is a fast track on the F1 GP". Then the probability of the sentence is given as:
$$
P = P(\text{<s>Monza is a fast track on the F1 GP</s>})
$$
Where `<s>` and `</s>` are predefined start and end tokens respectively.

We can then divide the sentence into tokens and use the chain rule of probability to make the calculations easier.
$$
P = P(\text{Monza|<s>}).P(\text{is|<s>Monza}).P(\text{a|<s>Monza is}).P(\text{fast|<s>Monza is a}) ... .P(\text{</s>|<s>Monza is a fast track on the F1 GP})
$$

Learning the joint probability distirbution is much more difficult and hence usually we have Markov Assumption (https://en.wikipedia.org/wiki/Discrete-time_Markov_chain). The main idea is the current token only depend on the previous token and you can ignore the history before the previous token. Given the Markov Assumption, we have:

$$
P(\text{is|<s>Monza}) = P(\text{is|Monza}) \\
P(\text{a|<s>Monza is}) = P(\text{a|is}) \\
P(\text{fast|<s>Monza is a}) = P(\text{fast|a}) \\
... \\
P(\text{</s>|<s>Monza is a fast track on the F1 GP}) = P(\text{</s>|GP})
$$

Hence, we can write the probability of the complete sentence as

$$
P(\text{<s>Monza is a fast track on the F1 GP</s>}) = \\P(\text{Monza|<s>}).P(\text{is|Monza}).P(\text{a|is})...P(\text{</s>|GP})
$$

Above we made 1st order Markov Assumption, giving us a **BiGram Language Model**. Similarly, we can make $(n-1)^{st}$ order Markov assumption, giving us  **n-gram Language Model**

In general using n-gram model, the probability of a sentence $s$ is given as:
$$
P(s) = \prod{P(w_n|w_{1} w_{2} w_{3} w_{4} ... w_{n-1})}
$$

## 1. Introduction to TF-IDF

**Term Frequency-Inverse Document Frequency (TF-IDF)** is a technique used to convert text documents into numerical feature vectors, capturing the importance of words within documents relative to a corpus (https://en.wikipedia.org/wiki/Tf%E2%80%93idf).

### Key Components:

- **Term Frequency (TF):** Measures how often a term appears in a document.
  $$
  TF(t, d) = \frac{\text{Number of times term } t \text{ appears in document } d}{\text{Total number of terms in document } d}
  $$

- **Inverse Document Frequency (IDF):** Measures how important a term is by considering the rarity of the term across all documents.
  $$
  IDF(t, D) = \log \left(\frac{\text{Total number of documents in the corpus}}{\text{Number of documents containing the term } t}\right)
  $$

- **TF-IDF Calculation:** Combines TF and IDF values to produce a weighted score for each term in a document.
  $$
  TF-IDF(t, d, D) = TF(t, d) \times IDF(t, D)
  $$

## 2. Introduction to K-Nearest Neighbors (KNN)

**K-Nearest Neighbors (KNN)** is a simple, non-parametric algorithm used for classification and regression. It predicts the class of a data point based on the classes of its nearest neighbors (https://en.wikipedia.org/wiki/K-nearest_neighbors_algorithm).

### Steps in KNN Classification:

1. **Distance Calculation:** Compute the distance between a test document and all training documents. The Euclidean distance is commonly used:
   $$
   d(x, y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
   $$

2. **Identify Neighbors:** Find the `k` closest training documents to the test document.

3. **Prediction:** Assign the most common class label among the `k` nearest neighbors to the test document.

```python
import re
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk #Added this line to enable downloading stopwords

nltk.download('stopwords') #Added this line to download stopwords

def extract_tfidf_features(train_df, test_df, text_column='text'):


    stop_words = set(stopwords.words('english'))

    def preprocess_text(text):

        text = text.lower() # converting the docs to lower case letters

        text = ' '.join([word for word in text.split() if word not in stop_words])
        return text # removing stop words from the docs

    train_docs = [preprocess_text(doc) for doc in train_df[text_column]]
    test_docs = [preprocess_text(doc) for doc in test_df[text_column]]

    vectorizer = TfidfVectorizer(
        max_features=5000,     # Limiting vocabulary size
        min_df=5,             # Ignoring terms below a minimum document frequency
        ngram_range=(1, 2),   # Considering both single words unigrams and bigrams
        sublinear_tf=True      # Using sublinear term frequency scaling
    )


    tfidf_matrix_train = vectorizer.fit_transform(train_docs) # extracting the tfidf vector representation of the training data
    tfidf_matrix_test = vectorizer.transform(test_docs) # extracting the tfidf vector representation of the test data
    token_to_index = vectorizer.vocabulary_ # Creating a mapping from tokens to their indices in the tfidf matrix

    return tfidf_matrix_train, tfidf_matrix_test, token_to_index

# applying the function on our dataset
tfidf_matrix_train, tfidf_matrix_test, token_to_index = extract_tfidf_features(train_df, test_df)

# 1. NLTK Documentations https://www.nltk.org/
# 2. sklearn tfidf documentation https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html
```

```python

import numpy as np
from collections import Counter

def euclidean_distance(vector1, vector2):
    #Calculating the Euclidean distance between two vectors
    return np.linalg.norm(vector1 - vector2)

def cosine_distance(vector1, vector2):
    #Calculating the cosine distance between two vectors
    dot_product = np.dot(vector1, vector2)
    norm_1 = np.linalg.norm(vector1)
    norm_2 = np.linalg.norm(vector2)
    return 1 - (dot_product / (norm_1 * norm_2))

def predict_knn(test_instance, train_data, train_labels, k, distance_metric):


    distances = []

    test_instance = test_instance.toarray()  # Converting to dense arrays
    train_data = train_data.toarray()  # Converting to dense arrays


    for i, train_instance in enumerate(train_data):
        if distance_metric == 'euclidean':
            dist = euclidean_distance(test_instance, train_instance)
        elif distance_metric == 'cosine':
            dist = cosine_distance(test_instance, train_instance)

        distances.append((dist, train_labels[i]))

    # Sorting the distances and picking out the labels of the k-nearest neighbors
    distances.sort(key=lambda x: x[0])
    neighbor_labels = [label for _, label in distances[:k]]

    # Finding the most common label among the k nearest neighbors
    predicted_label = Counter(neighbor_labels).most_common(1)[0][0]
    return predicted_label

#  Citations
#  1. https://en.wikipedia.org/wiki/Euclidean_distance to learn about euclidean distance
#  2. https://en.wikipedia.org/wiki/Cosine_similarity to learn about cosine similarities between vectors
#  3. https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import pandas as pd

def predict_knn_batch(test_data, train_data, train_labels, k, distance_metric):
  # To get predictions on a particular batch and then appending the prediction in a predictions[], at the end this will contain all the predictions for the dataset

    predictions = []
    for test_instance in test_data:
        prediction = predict_knn(test_instance, train_data, train_labels, k, distance_metric)
        predictions.append(prediction)
    return predictions

def evaluate_knn_with_batching(train_data, train_labels, test_data, test_labels,
                              k_values, distance_metric, batch_size):


    results = {'k': [], 'accuracy': [], 'precision': [], 'recall': [], 'f1_score': []}

    for k in k_values:
        all_predictions = []
        for i in range(0, (test_data).shape[0], batch_size):
            batch_predictions = predict_knn_batch(test_data[i:i + batch_size],
                                                  train_data, train_labels, k, distance_metric)
            all_predictions.extend(batch_predictions)

        # Calculate and store metrics for the current k
        results['k'].append(k)
        results['accuracy'].append(accuracy_score(test_labels, all_predictions))
        results['precision'].append(precision_score(test_labels, all_predictions, average='weighted'))
        results['recall'].append(recall_score(test_labels, all_predictions, average='weighted'))
        results['f1_score'].append(f1_score(test_labels, all_predictions, average='weighted'))

    return pd.DataFrame(results)

def visualize_knn_performance(results_df):

    fig, axs = plt.subplots(2, 1, figsize=(10, 12))

    # Plotting accuracy vs k
    axs[0].plot(results_df['k'], results_df['accuracy'], marker='o', label='Accuracy')
    axs[0].set_title('Accuracy vs k')
    axs[0].set_xlabel('k')
    axs[0].set_ylabel('Accuracy')
    axs[0].grid(True)

    # Plotting F1-score vs k
    axs[1].plot(results_df['k'], results_df['f1_score'], marker='o', label='F1-score')
    axs[1].set_title('F1-score vs k')
    axs[1].set_xlabel('k')
    axs[1].set_ylabel('F1-score')
    axs[1].grid(True)

    plt.tight_layout()
    plt.show()

    #  Citations
    #  1. https://scikit-learn.org/stable/api/sklearn.metrics.html for metrics
    #  2. https://hevodata.com/learn/python-batch-processing/ for batch processing techniques
```

```python
!pip install annoy
import annoy
from annoy import AnnoyIndex
from scipy.sparse import csr_matrix

# Using Annoy to speed up k-NN search with sparse matrix and batch processing
def knn_with_annoy_sparse(train_vectors, train_labels, test_vectors, k, num_trees=10, batch_size=100):
    f = train_vectors.shape[1]

    # Building the Annoy index
    annoy_index = AnnoyIndex(f, 'euclidean')  # Using Euclidean distance
    for i in range(train_vectors.shape[0]):
        annoy_index.add_item(i, train_vectors[i].toarray().ravel())  # Converting each row to dense only when needed

    annoy_index.build(num_trees)

    # Query for nearest neighbors in batches

    predictions = []

    num_batches = (test_vectors.shape[0] // batch_size) + 1
    for batch in range(num_batches):
        start = batch * batch_size
        end = min((batch + 1) * batch_size, test_vectors.shape[0])

        for i in range(start, end):
            test_vector_dense = test_vectors[i].toarray().ravel()  # Converting one test vector at a time
            indices = annoy_index.get_nns_by_vector(test_vector_dense, k)
            neighbor_labels = [train_labels[idx] for idx in indices]
            predictions.append(max(set(neighbor_labels), key=neighbor_labels.count))


    return predictions

k = 20
batch_size = 50
annoy_predictions = knn_with_annoy_sparse(train_vectors, train_labels, test_vectors, k, num_trees=50, batch_size=batch_size)

# This was very fast compared to normal knn. This cell took 36 seconds.
# These results are obtained quickly due to Annoy's optimized querying capabilities.

#  Citations
# 1. https://github.com/spotify/annoy
# 2. https://docs.scipy.org/doc/scipy/reference/sparse.html to study about sparse matrices
```

```python

import time
from sklearn.neighbors import KDTree
from sklearn.decomposition import TruncatedSVD

def knn_predict_with_kdtree(train_features, train_labels, test_features, k):

    start_time = time.time()
    tree = KDTree(train_features, leaf_size=40)  # Building KD-Tree
    _, indices = tree.query(test_features, k=k)  # Query for nearest neighbors

    predictions = []
    for neighbor_indices in indices:
        neighbor_labels = [train_labels[idx] for idx in neighbor_indices]
        # Predicting the most frequent class among neighbors
        predictions.append(max(set(neighbor_labels), key=neighbor_labels.count))

    runtime = time.time() - start_time
    print(f"KD-Tree k-NN runtime: {runtime:.4f} seconds")
    return predictions

def reduce_dimensionality_with_svd(train_data, test_data, n_components=50):
  #   reducing the dimentionality to 50 components
    svd = TruncatedSVD(n_components=n_components)
    train_data_reduced = svd.fit_transform(train_data)
    test_data_reduced = svd.transform(test_data)
    return train_data_reduced, test_data_reduced


train_vectors, test_vectors = reduce_dimensionality_with_svd(
    train_vectors, test_vectors, n_components=50
)

k_neighbors = 20 # chosen 20 nearest neighbours for the predictions
kd_tree_svd_predictions = knn_predict_with_kdtree(
    train_vectors, train_labels, test_vectors, k=k_neighbors
)

#  I have reduced the dimensions of the data using svd to 50 components as pca doesnot handle sparse matrix so I have chosen svd to do that job
#  This was the fastest this cell took only 2 seconds to run.

# Citations
# 1. https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KDTree.html for KD tree
# 2. https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.TruncatedSVD.html for truncated SVD
# 3. https://medium.com/nerd-for-tech/dimensionality-reduction-techniques-pca-lca-and-svd-f2a56b097f7c for learning about dimensioanlity reduction
```

### 1. Introduction to K-Means Clustering

K-Means is a popular clustering algorithm used to partition data into k distinct clusters. Each cluster is represented by its centroid, which is the mean of all points assigned to that cluster.

**Steps in K-Means Clustering:**
- **Initialization:** Choose a value of k (this corresponds to number of clusters). This is a hyper-parameter that you need to experiment with.
- **Assignment:** Assign each data point to the nearest centroid.
- **Update:** Recalculate centroids as the mean of the points assigned to each cluster.
- **Iteration:** Repeat the assignment and update steps until centroids no longer change significantly.

```python
def cluster_and_predict(train_features, train_labels, test_features, n_clusters):

    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(train_features)

    train_cluster_assignments = kmeans.predict(train_features)
    test_cluster_assignments = kmeans.predict(test_features)
    return train_cluster_assignments, test_cluster_assignments

# using sklearn's kmeans model and fit it on our dataset

# Citations
# 1. https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
```

```python
# Mapping Clusters to Categories
def map_clusters_to_labels(train_labels, train_cluster_assignments, test_cluster_assignments):

  cluster_to_category = {}
  for cluster in range(n_clusters):
      cluster_indices = train_cluster_assignments == cluster
      most_common_category = train_labels[cluster_indices].mode()[0]
      cluster_to_category[cluster] = most_common_category

  test_predicted_categories = [cluster_to_category[c] for c in test_cluster_assignments]
  return test_predicted_categories

# as we are getting the clusters and now we have to retrieve labels from each clusters we are mapping the clusters to the labels of that label which is most frequent in that cluster

# Citations
# 1. https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
```

```python

def evaluate_clustering(true_labels, predicted_labels):
  accuracy = accuracy_score(true_labels, predicted_labels)
  precision = precision_score(true_labels, predicted_labels, average='weighted')
  recall = recall_score(true_labels, predicted_labels, average='weighted')
  f1 = f1_score(true_labels, predicted_labels, average='weighted')

  print(f"Accuracy: {accuracy:.4f}")
  print(f"Precision: {precision:.4f}")
  print(f"Recall: {recall:.4f}")
  print(f"F1-score: {f1:.4f}")
  return accuracy, precision, recall, f1

# Evluation of the cluster labels
```

```python
# Experimenting with Different k values
k_values_to_test = [2, 6, 10, 13,16,20]
accuracies = []
f1_scores = []

for n_clusters in k_values_to_test:
    print(f"\nEvaluating for k = {n_clusters}")
    train_clusters, test_clusters = cluster_and_predict(
        train_vectors, train_df['category'], test_vectors, n_clusters
    )

    test_predicted_categories = map_clusters_to_labels(
        train_df['category'], train_clusters, test_clusters
    )

    acc, _, _, f1 = evaluate_clustering(test_df['category'], test_predicted_categories)
    accuracies.append(acc)
    f1_scores.append(f1)

#  Visualization
plt.figure(figsize=(8, 5))
plt.plot(k_values_to_test, accuracies, label='Accuracy', marker='o')
plt.plot(k_values_to_test, f1_scores, label='F1-Score', marker='x')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Score')
plt.title('K-Means Clustering Performance')
plt.xticks(k_values_to_test)
plt.legend()
plt.grid(True)
plt.show()
```

### **Background: Singular Value Decomposition (SVD)**

SVD is a matrix factorization technique in linear algebra. It decomposes a matrix M  into three matrices:  
$M = U \Sigma V^{*}$

Where:
- $M$ is the original matrix.
- $U$ is an $m \times m$  unitary matrix.
- $\Sigma$ is an $m \times n$  diagonal matrix with non-negative real numbers (singular values).
- $V^{*}$ is the conjugate transpose of $V$ , an $n \times n$ unitary matrix.

This decomposition is useful in reducing the dimensionality of data while retaining important features, making it a powerful tool for tasks like latent semantic analysis (LSA) (https://en.wikipedia.org/wiki/Latent_semantic_analysis) in Natural Language Processing (NLP). You can also watch this nice video lecture series on SVD: https://www.youtube.com/watch?v=gXbThCXjZFM

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def clean_text(text, remove_stopwords=True, stemming=False):
    if isinstance(text, str):
        text = re.sub(r'[^a-zA-Z\s]', '', text) # used regex for splitting the words of the english language on the basis of white spaces
        text = ' '.join(text.split())  # Normalize whitespace

        if remove_stopwords: # Removing stop words
            text = ' '.join([word for word in text.split() if word.lower() not in stop_words])

        if stemming: # stemming
            stemmer = nltk.PorterStemmer()
            text = ' '.join([stemmer.stem(word) for word in text.split()])

        return text.lower() # converting the text into lower case alphabets
    return ''

# Citations
# 1. https://docs.python.org/3/library/re.html for learning about regex
# 2. https://www.nltk.org/
# 3. https://www.nltk.org/howto/stem.html for stemming
# 4. https://www.nltk.org/nltk_data/ for stopwords
```

```python
from collections import defaultdict  # Importing defaultdict from collections to create a dictionary that assigns default values
from scipy.sparse import lil_matrix  # Importing lil_matrix for creating a sparse matrix to save memory space
import numpy as np


vocab = defaultdict(lambda: len(vocab))  # assigning an increasing integer index for each new word
vocab_inv = {}  # A reverse lookup dictionary to map index to word
corpus = []  # A list to store the corpus as a collection of word indices

def build_vocabulary_and_corpus(text_data):
    # Looping through each document in the text data
    for document in text_data:
        tokenized_document = document.split()  # Spliting document into individual words (tokens)
        document_indices = []  # Store the word indices for each document

        # Looping through each token in the document
        for token in tokenized_document:
            token_index = vocab[token]  # index of the word from the vocab (assigning a new one if not present)
            vocab_inv[token_index] = token  # Adding the reverse lookup (index to word) in vocab_inv
            document_indices.append(token_index)  # Appending the index to the document_indices list

        # Appending the list of word indices for the document to the corpus
        corpus.append(document_indices)

    return vocab, vocab_inv, corpus

vocabulary, inverse_vocabulary, corpus_data = build_vocabulary_and_corpus(train_df_sampled['lemmatized_text'])

# size of the vocabulary (number of unique words)
vocabulary_size = len(vocabulary)  # This gives the total number of unique words in the corpus


# Citations
# 1. https://docs.python.org/3/library/collections.html#collections.defaultdict for learning about vocab
# 2. https://www.scipy.org/ for sparse matrix
# 3. https://numpy.org/ for numpy array
```

```python

import numpy as np
from scipy.sparse import lil_matrix

def build_word_context_matrix(corpus, vocabulary_size, window_size):


  word_context_matrix = lil_matrix((vocabulary_size, vocabulary_size), dtype=np.int32)# creating a sparse matrix of size (vocab_size)*(vocab_size)

  for document in corpus:
    for i, target_word in enumerate(document):
      # context window boundaries
      #  we have to take 2 words before and 2 words after a particular words into account for making the context
      left_boundary = max(0, i - window_size // 2)
      right_boundary = min(len(document), i + window_size // 2 + 1)

      # Extracting context words
      context_words = document[left_boundary:i] + document[i+1:right_boundary]

      # Incrementing co-occurrence counts in the matrix
      for context_word in context_words:
        word_context_matrix[target_word, context_word] += 1

  return word_context_matrix

#  checking for different context sized context window

k_5 = 5  # Neighborhood size (context window size)
k_10 = 10  # Neighborhood size (context window size)
k_15 = 15  # Neighborhood size (context window size)
word_context_matrix_5 = build_word_context_matrix(corpus, vocabulary_size, k_5)
word_context_matrix_10 = build_word_context_matrix(corpus, vocabulary_size, k_10)
word_context_matrix_15 = build_word_context_matrix(corpus, vocabulary_size, k_15)

# Citations
# 1. https://nlp.stanford.edu/IR-book/ for learning about information retrieval
```

```python
from sklearn.decomposition import TruncatedSVD
def perform_svd(word_context_matrix, num_components=100):

  svd = TruncatedSVD(n_components=num_components)
  U = svd.fit_transform(word_context_matrix)
  Sigma = svd.singular_values_
  Vt = svd.components_

  print("U shape:", U.shape)
  print("Sigma shape:", Sigma.shape)
  print("Vt shape:", Vt.shape)

  return U, Sigma, Vt

#  Reducing the dimensionality of the word_context_matrix to 100 dimensions using truncated svd
num_dimensions = 100  # Desired number of dimensions for word embeddings
word_embeddings_5, singular_values_5, Vt_5 = perform_svd(word_context_matrix_5, num_dimensions)
word_embeddings_10, singular_values_10, Vt_10 = perform_svd(word_context_matrix_10, num_dimensions)
word_embeddings_15, singular_values_15, Vt_15 = perform_svd(word_context_matrix_15, num_dimensions)

# Citations
# 1. https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.TruncatedSVD.html#sklearn.decomposition.TruncatedSVD
```

```python

import numpy as np
import matplotlib.pyplot as plt

def analyze_svd_variance(singular_values):

  explained_variance = np.square(singular_values) / np.sum(np.square(singular_values))
  cumulative_variance = np.cumsum(explained_variance)

  plt.figure(figsize=(8, 3))
  plt.plot(cumulative_variance, marker='o', linestyle='-', color='b')
  plt.title("Cumulative Variance Explained by Singular Values")
  plt.xlabel("Number of Singular Values (r)")
  plt.ylabel("Cumulative Variance Explained")
  plt.grid(True)
  plt.show()

analyze_svd_variance(singular_values_5)
analyze_svd_variance(singular_values_10)
analyze_svd_variance(singular_values_15)
```

```python

import numpy as np

def get_truncated_embeddings(U, Sigma, r):


  Sigma_r = np.diag(Sigma[:r])  # Create a diagonal matrix with top r values
  Sigma_r_inv = np.linalg.inv(Sigma_r) # Invert the diagonal matrix
  word_embeddings = np.dot(U[:, :r], Sigma_r_inv)

  print("Word embeddings shape:", word_embeddings.shape)
  return word_embeddings

# So now reducing the dimension for faster computation
r_5=15
r_10=15
r_15=15
word_embeddings_5 = get_truncated_embeddings(word_embeddings_5, singular_values_5, r_5)
word_embeddings_10 = get_truncated_embeddings(word_embeddings_10, singular_values_10, r_10)
word_embeddings_15 = get_truncated_embeddings(word_embeddings_15, singular_values_15, r_15)

# Citations
# 1. https://numpy.org/ fro  or manipulating the matrices
```

```python

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

#  Calculating cosine similarity between two words with the help of word embeddings
def calculate_cosine_similarity(word_pairs, vocabulary, word_embeddings):


  word_to_index = {word: idx for idx, word in enumerate(vocabulary)}

  for word1, word2 in word_pairs:
    try:
      idx1 = word_to_index[word1]
      idx2 = word_to_index[word2]

      vec1 = word_embeddings[idx1].reshape(1, -1)
      vec2 = word_embeddings[idx2].reshape(1, -1)

      similarity = cosine_similarity(vec1, vec2)[0][0]
      print(f"Cosine similarity between '{word1}' and '{word2}': {similarity:.4f}")
# If the given word is not in the vocab then we are not able to get the similarities so to handle that case my solution will give the below output
    except KeyError as e:
      print(f"Word not found in vocabulary: {e}")

# for checking purpose I have used these examples
word_pairs = [
    ('car', 'automobile'),
    ('dog', 'cat'),
    ('animal', 'dog'),
    ('table', 'furniture')
]
print("Cosine similarities by taking the context window as 5")
calculate_cosine_similarity(word_pairs, vocabulary, word_embeddings_5)
print("Cosine similarities by taking the context window as 10")
calculate_cosine_similarity(word_pairs, vocabulary, word_embeddings_10)
print("Cosine similarities by taking the context window as 15")
calculate_cosine_similarity(word_pairs, vocabulary, word_embeddings_15)

# Citations
# 1. https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.cosine_similarity.html
```
